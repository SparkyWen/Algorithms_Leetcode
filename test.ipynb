```python
import sys

def parse_floor(s):
    if not s:
        raise ValueError("cant be null")
    if s[0] =='F':
        body = s[1:]
        if body.endwith("A"):
            return int(body[:-1]) + 1
        else:
            return int(body)
    elif s[0] == 'B':
        return 1 - int(s[1:])
    else:
        raise ValueError("invalid floor")
```
